```python
import numpy as np
import pandas as pd

data = pd.read_csv('bar_events_with_embeddings.csv',index_col=0)
data2 = pd.read_csv("funcheap_events_with_embeddings.csv",index_col=0)

# concatenate data frames vertically
df = pd.concat([data, data2], axis=0)
```

```python
def find_neighborhood(venue):
    prompt = "Given the following venue, provide the SF neighborhood that this venue is located in." + '\n' + 'Venue: ' + venue
    return prompt
```

```python
import openai
import os

openai.apikey = os.getenv("OPENAI_API_KEY")

venue_description = {}
for index, row in df.iterrows():
    if row['venue'] not in venue_description:
        response = openai.Completion.create(
            model="text-curie-001",
            prompt=find_neighborhood(row['venue']),
            temperature=0.0,
            max_tokens = 50,
        )
        response_text = response['choices'][0]['text'].strip()
        venue_description[row['venue']] = response_text
        df.loc[index, 'neighborhood'] = response_text

    else:
        df.loc[index, 'neighborhood'] = venue_description[row['venue']]  
```

```python
df['combined'] = df['combined'] + '\n' + df['neighborhood']
```

## Embed User and Events Data

```python
df_users = pd.read_csv('sf_user_data.csv',index_col=0)
```

```python
df_users.head(2)
```

```python
import pandas as pd
import tiktoken
from openai.embeddings_utils import get_embedding
import os
import openai

openai.apikey = os.getenv("OPENAI_API_KEY")
# embedding model parameters
embedding_model = "text-embedding-ada-002"
embedding_encoding = "cl100k_base"  # this the encoding for text-embedding-ada-002
max_tokens = 5000  # the maximum for text-embedding-ada-002 is 8191
```

```python
df_users = df_users.applymap(str)
encoding = tiktoken.get_encoding(embedding_encoding)
df_users["combined"] = (
    "Gender: " + df_users.gender.str.strip() + "; Age: " + df_users.age.str.strip() + 
    "; Event Preferences: " + df_users.event_preferences.str.strip() +  "; Venue Preferences: " + df_users.venue_preferences.str.strip() + 
    "; Price Range Preferences: " + df_users.price_range_pref.str.strip() +  "; Time of Day Preferences: " + df_users.venue_preferences.str.strip() + 
    "; Time of Day Preferences: " + df_users.time_of_day_pref.str.strip() + "; Neighborhoods of Interest: " + df_users.neighborhoods_of_interest.str.strip()
)

df_users["n_tokens"] = df_users.combined.apply(lambda x: len(encoding.encode(x)))
df_users["embedding"] = df_users.combined.apply(lambda x: get_embedding(x, engine=embedding_model))

```

```python
df["embedding"] = df.combined.apply(lambda x: get_embedding(x, engine=embedding_model))
```

## Match Users to Events

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


user_embeddings = np.array(df_users['embedding'].tolist())
event_embeddings = np.array(df['embedding'].tolist())

similarity_matrix = cosine_similarity(user_embeddings, event_embeddings)
```

```python
N = 10

recommendations = {}

for i, user_id in enumerate(df_users['email']):
    top_N = np.argsort(similarity_matrix[i])[::-1][:N]
    recommended_events = df.iloc[top_N]['event'].tolist()
    recommendations[user_id] = recommended_events
```

```python
for k,v in recommendations.items():
    recommendations[k] = set(v)

for k,v in recommendations.items():
    print(k,v)
    break
```
