```python
import pandas as pd

user_df = pd.read_csv('MOCK_DATA.csv')
```

```python
user_df
```

```python
user_df = user_df.drop(['id', 'latitude', 'longitude'], axis=1)
```

```python
import random 
for i in range(1000):
    val = random.uniform(37.643, 37.8272)
    user_df.loc[i, 'latitude'] = val
```

```python
for i in range(1000):
    val = random.uniform(122.3927, 122.5283)
    user_df.loc[i, 'longitude'] = val
```

```python
interests = ['Live music', 'DJ performances', 'Dancing', 'Karaoke', 'Comedy shows', 'Trivia nights', 'Art exhibits', 'Film screenings', 'Wine tasting', 'Craft beer tasting', 'Mixology classes', 'Food events', 'Sports viewing parties', 'Networking events', 'Tech events']
venue_types = ['Speakeasy Bar', 'Dive Bar', 'Nightclubs', 'Music Venue', 'Comedy Club', 'Theaters', 'Art Gallery', 'Film theaters', 'Wineries', 'Breweries', 'Cocktail Bars', 'Sports Bars', 'Restaurants with Live Music', 'Rooftop bars', 'Convention Centers']



user_df['event_preferences'] = [random.sample(interests, k=5) for _ in range(len(user_df))]
# Add a new column to the DataFrame with random venue types
user_df['venue_preferences'] = [random.sample(venue_types, k=5) for _ in range(len(user_df))]
```

```python
price_range_pref = ['free', '$10-$20', '$20-$50', '$50-$100', 'over $100']
time_of_day_pref = ['daytime', 'evening', 'night']
neighborhoods_of_interest = ['Financial District', 'Nob Hill', 'Pacific Heights', 'Russian Hill', 'North Beach', 'The Mission', 'Castro', 'Haight-Ashbury', 'SoMa']

user_df['price_range_pref'] = [random.sample(price_range_pref, k=random.randint(1,len(price_range_pref))) for _ in range(len(user_df))]
user_df['time_of_day_pref'] = [random.sample(time_of_day_pref, k=random.randint(1,len(time_of_day_pref))) for _ in range(len(user_df))]
user_df['neighborhoods_of_interest'] = [random.sample(neighborhoods_of_interest, k=random.randint(1,len(neighborhoods_of_interest))) for _ in range(len(user_df))]
```

## Embed User Data

```python
import pandas as pd
import tiktoken
from openai.embeddings_utils import get_embedding
import os
import openai

openai.apikey = os.getenv("OPENAI_API_KEY")
# embedding model parameters
embedding_model = "text-embedding-ada-002"
embedding_encoding = "cl100k_base"  # this the encoding for text-embedding-ada-002
max_tokens = 5000  # the maximum for text-embedding-ada-002 is 8191
```

```python
df_users = user_df.applymap(str)
encoding = tiktoken.get_encoding(embedding_encoding)
user_df["combined"] = (
    "Gender: " + df_users.gender.str.strip() + "; Age: " + df_users.age.str.strip() + 
    "; Event Preferences: " + df_users.event_preferences.str.strip() +  "; Venue Preferences: " + df_users.venue_preferences.str.strip() + 
    "; Price Range Preferences: " + df_users.price_range_pref.str.strip() +  "; Time of Day Preferences: " + df_users.venue_preferences.str.strip() + 
    "; Time of Day Preferences: " + df_users.time_of_day_pref.str.strip() + "; Neighborhoods of Interest: " + df_users.neighborhoods_of_interest.str.strip()
)

user_df["n_tokens"] = user_df.combined.apply(lambda x: len(encoding.encode(x)))
user_df["embedding"] = user_df.combined.apply(lambda x: get_embedding(x, engine=embedding_model))
```

```python
user_df.to_csv('sf_users.csv')
```
